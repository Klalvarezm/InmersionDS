# 1. Promedio de area de todos los barrios

Saber cuantos barrios hay

```python
len(inmuebles.Barrio.unique())
```

Esta lista guarda los nombres de cada barrio

```python
ListaBarrios=inmuebles.Barrio.unique()
```

Basicamente lo que se hizo fue extraer el promedio de area de barrio a barrio y colocarlo en un diccionario, incluyendo nombre del barrio y promedio de area para cada uno

```python

dictPromedio={}
i=0
#inmueblesChico=(inmuebles.Barrio=='Chico Reservado')
#chico_reservado=inmuebles[inmueblesChico]
#chico_reservado.Area.mean()
for barrio in ListaBarrios:
  currentBarrio=inmuebles[(inmuebles.Barrio==barrio)]
  dictPromedio.update({barrio:currentBarrio.Area.mean()})
dictPromedio

```

Esa informacion se pasa de un diccionario a un dataframe de pandas.

Donde se organiza de mayor a menor area.

```python
promedioBarrio=pd.DataFrame(dictPromedio.items(),columns=['Nombre','Promedio Area'])
promedioBarrio=promedioBarrio.sort_values(by=['Promedio Area'],ascending=False)
promedioBarrio
```

Se imprime la info en una grafica 

```python

promedioBarrio.plot.bar()
```

Top 10 barrios con mas area promedio.

```python
promedioBarrio.axes
```

```python
promedioBarrio.head(10).plot.bar("Nombre","Promedio Area")
```

# Usando Metodo correcto "Group By"

```python

promedioAreaPorBarrio= inmuebles.groupby('Barrio').Area.mean()
promedioAreaPorBarrio = pd.DataFrame(promedioAreaPorBarrio).reset_index()
promedioAreaPorBarrio.columns = ['Barrio', 'Promedio Area']
promedioAreaPorBarrio=promedioAreaPorBarrio.sort_values(by=['Promedio Area'],ascending=False)
promedioAreaPorBarrio
```

```python
promedioAreaPorBarrio.head(10).plot.bar('Barrio','Promedio Area')
```

## Promedio de costo por barrio

Limpiando el string de valor

```python
inmuebles['Valor']=inmuebles['Valor'].str.replace("$", "")
inmuebles['Valor']=inmuebles['Valor'].str.replace(" ", "")
inmuebles['Valor']=inmuebles['Valor'].str.replace(".", "")
inmuebles['Valor']=pd.to_numeric(inmuebles['Valor'])

```

Conversion Exitosa de valor

```python
inmuebles
```

#Uso este para que el dataframe muestre el valor completo y no notacion cientifica

```python
pd.options.display.float_format = '{:.2f}'.format
```

```python


promedioCostoPorBarrio= inmuebles.groupby('Barrio').Valor.mean()
promedioCostoPorBarrio = pd.DataFrame(promedioCostoPorBarrio).reset_index()
promedioCostoPorBarrio.columns = ['Barrio', 'Promedio Precio']
promedioCostoPorBarrio=promedioCostoPorBarrio.sort_values(by=['Promedio Precio'],ascending=False)
promedioCostoPorBarrio
```
